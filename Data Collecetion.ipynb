```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm

def generate_tgl(tgl_awal, tgl_akhir, bulan, tahun):
    tgl_ambil=[]
    if(bulan<10):
        bulan='-0'+str(bulan)
    else:
        bulan='-'+str(bulan)
    for i in range(tgl_awal,tgl_akhir+1):
        if i<10:
            i='-0'+str(i)
        else:
            i='-'+str(i)
        tgl_ambil.append(str(tahun)+bulan+i)
    return tgl_ambil

def scraping_kompas(key, tgl_ambil):
    key_tgl=['https://indeks.kompas.com/?site={}&date={}'.format(key, j) for j in tgl_ambil]
    list1=[]
    for url in tqdm(key_tgl, total=len(key_tgl)):
        req=requests.get(url)
        soup=BeautifulSoup(req.text, 'lxml')
        kumpulan_page=soup.find_all('div',class_='article__asset')
        link_page=[kumpulan_page[i].a['href'] for i in range(len(kumpulan_page))]
        # print('ini url page', url)
        count=0
        for i in link_page:
            temp={}
            count+=1
            i=i+'?page=all' #to make view all page
            req_berita=requests.get(i)
            soup_berita=BeautifulSoup(req_berita.text, 'lxml')
            temp['judul']=soup_berita.find_all('title')[0].text
            temp['isi']=soup_berita.find_all('div', {'class':'read__content'})[0].text.strip().replace('\n' , ' ')
            temp['tanggal']=soup_berita.find_all('div', {'class':'read__time'})[0].text.split(',')[0].split('-')[1].replace(' ',"")
            temp['penulis']=soup_berita.find_all('div',{'class':'read__credit__item'})[0].find('a').text
            temp['kategori']=soup_berita.find_all('span',{'itemprop':'name'})[2].text
            # print(berita)
            # Link += j.find("h3",attrs={'class':'article__title article__title--medium'}).find('a')['href'].strip()
            temp['photo']=soup_berita.find("div", attrs={'class':'photoCarousel-img'}).img['data-src']
            # temp['komen']=soup_berita.find('span', {'class':'comment__read__text'}).find('a')['herf']
            # temp['photo']=soup_berita.find_all('div', {'class':'photo__wrap'})
            list1.append(temp)
            # print(title, tgl, photo)
    return list1

if __name__ == "__main__":
    range_tgl=generate_tgl(1,23,4,2022)
    key='tekno'
    df=pd.DataFrame(scraping_kompas(key, range_tgl))
    df.to_excel('/kompas_april22.xlsx')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm

def generate_tgl(tgl_awal, tgl_akhir, bulan, tahun):
    tgl_ambil=[]
    if(bulan<10):
        bulan='-0'+str(bulan)
    else:
        bulan='-'+str(bulan)
    for i in range(tgl_awal,tgl_akhir+1):
        if i<10:
            i='-0'+str(i)
        else:
            i='-'+str(i)
        tgl_ambil.append(str(tahun)+bulan+i)
    return tgl_ambil

def scraping_kompas(tgl_ambil):
    key_tgl=['https://index.sindonews.com/index/612?t={}'.format(j) for j in tgl_ambil]
    list1=[]
    for url in tqdm(key_tgl, total=len(key_tgl)):
        req=requests.get(url)
        soup=BeautifulSoup(req.text, 'lxml')
        
        kumpulan_page=soup.find_all('div',class_='indeks-title')
        link_page=[kumpulan_page[i].a['href'] for i in range(len(kumpulan_page))]
        # print('ini url page', link_page)
        count=0
        for i in link_page:
            temp={}
            count+=1
            i=i+'?showpage=all' #to make view all page
            req_berita=requests.get(i)
            soup_berita=BeautifulSoup(req_berita.text, 'lxml')
            temp['judul']=soup_berita.find_all('title')[0].text
            temp['isi']=soup_berita.find_all('div', {'class':'desc-artikel-detail vidy-embed'})[0].text.strip().replace('\n' , ' ')
            temp['tanggal']=soup_berita.find_all('div', {'class':'date-detail'})[0].text.split(',')[1].split('-')[0]
            temp['penulis']=soup_berita.find_all('div',{'class':'auth-detail'})[0].find('a').text
            temp['kategori']=soup_berita.find_all('div',{'class':'sin-breadcrumb sin-clearfix'})[0].find_all('li')[1].text
            # print(berita)
            temp['photo']=soup_berita.find_all('div',{'class':'img-detail'})[0].find('img').get('src')
            list1.append(temp)
            # print(title, tgl, photo)
    return list1

if __name__ == "__main__":
    range_tgl=generate_tgl(1,31,4,2022) #(tgl_awal, tgl_akhir, bln, thn)
#     key='tekno'
    df_sindo=pd.DataFrame(scraping_kompas(range_tgl))
    df_sindo
    df_sindo.to_excel('/sindo_april22.xlsx')
```

```python
#import modules
import pandas as pd
import glob

# path of the folder
path = '/content/sindo'

# reading all the excel files
filenames = glob.glob(path + "/*.xlsx")
print('File names:', filenames)

# initializing empty data frame
finalexcelsheet = pd.DataFrame()

# to iterate excel file one by one
# inside the folder
for file in filenames:

	# combining multiple excel worksheets
	# into single data frames
	df = pd.concat(pd.read_excel(
	file, sheet_name=None), ignore_index=True, sort=False)

	# appending excel files one by one
	finalexcelsheet = finalexcelsheet.append(
	df, ignore_index=True)

# to print the combined data
print('Final Sheet:')
display(finalexcelsheet)

finalexcelsheet.to_excel(r'TeachSpace_Final.xlsx', index=False)
```
